```python
import pandas as pd
```

```python
df = pd.read_csv('../input/hpa-processed-train-dataframe-with-cellwise-rle/train_df_w_masks_shape_bboxes.csv')
```

```python
df
```

```python
for i in range(df.shape[0]):
    a = df.loc[i,'ID']
    f = open(f'{a}.txt', 'w')
    b = list(set(df.loc[i, 'Label'].split('|')))
    b_len = len(b)
    c = df.loc[i,'bboxes'][2:-2].split('), (')
    if c == ['']:
        f.close()
        continue       
    c_len = len(df.loc[i,'bboxes'][2:-2].split('), ('))
    for j in range(b_len):
        for k in range(c_len):
            d = c[k].split(', ')
            f.write(b[j]+' '+str((float(d[0]) + float(d[2])) / 2 / df.loc[i,'dimension']) +' '+str((float(d[1]) + float(d[3])) / 2 / df.loc[i,'dimension'])+' '+str((float(d[2]) - float(d[0])) / df.loc[i,'dimension'])+' '+str((float(d[3]) - float(d[1])) / df.loc[i,'dimension'])+' ')
            f.write('\n')

    f.close()
```
